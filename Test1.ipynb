```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Resources/oecd_gender_converted.csv')
```

```python
df
```

```python
df["Practice"].plot.box()
plt.show()
```

```python
df["Practice"].describe()
```

```python
bin_labels_3 = ["1","2","3"]
results, bin_edges = pd.qcut(df['Practice'],
                            q=[0, 1/3, 2/3, 1],
                            labels=bin_labels_3,
                            retbins=True)

results_table = pd.DataFrame(zip(bin_edges, bin_labels_3),
                            columns=['Threshold', 'Tier'])
```
